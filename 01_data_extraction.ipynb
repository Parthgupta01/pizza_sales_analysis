```python
import pymysql

# MySQL Database Credentials
host = "localhost"
user = "root"
password = "Parth@2005"
database = "pizza_sales"

# Establish Connection
try:
    connection = pymysql.connect(host=host, user=user, password=password, database=database)
    print(" Successfully connected to MySQL database!")
except Exception as e:
    print(" Connection failed:", e)
```

```python
import pandas as pd
# List of tables to fetch
tables = ["orders", "order_details", "pizzas", "pizza_types"]

# Loop through each table and save as CSV
for table in tables:
    query = f"SELECT * FROM {table};"
    cursor = connection.cursor()
    cursor.execute(query)

    # Fetch column names
    columns = [col[0] for col in cursor.description]

    # Fetch data & convert to Pandas DataFrame
    data = cursor.fetchall()
    df = pd.DataFrame(data, columns=columns)

    # Save Data to CSV
    df.to_csv(f"{table}.csv", index=False)
    print(f"{table}.csv saved successfully!")

    cursor.close()

# Close Connection
connection.close()
print(" MySQL Connection Closed!")
```

```python
# Load CSV Files
orders = pd.read_csv("orders.csv")
order_details = pd.read_csv("order_details.csv")
pizzas = pd.read_csv("pizzas.csv")
pizza_types = pd.read_csv("pizza_types.csv")

print("All datasets loaded successfully!")
```

```python
# Orders Table: Agar 'order_date' missing hai toh drop kar do
orders.dropna(subset=['date'], inplace=True)

# Pizzas Table: Agar 'price' missing hai toh uska median bhar do
pizzas['price'].fillna(pizzas['price'].median())
```

```python
# Remove Duplicates
orders.drop_duplicates(inplace=True)
order_details.drop_duplicates(inplace=True)
pizzas.drop_duplicates(inplace=True)
pizza_types.drop_duplicates(inplace=True)
```

```python
orders['date'] = pd.to_datetime(orders['date'])
```

```python
orders['day_of_week'] = orders['orddate'].dt.day_name()
```

```python
orders['hour'] = orders['time'].astype(str).str.split(' ').str[-1].str.split(':').str[0].astype(int)
```
